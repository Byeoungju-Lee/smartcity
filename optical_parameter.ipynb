```python
import numpy as np
import math
import pandas as pd
from itertools import combinations, product
import random
```

```python
def lorentzian( wavenumbers, w_0, w_p, gamma ):
    return wavenumbers * gamma * pow(w_p,2) / ( (w_0*gamma)**2 + ( pow(wavenumbers,2) - pow(w_0,2) )**2)

def multi_lorentz( wavenumbers, params ):
    #off = params[0]
    #paramsRest = params[1:]
    paramsRest = params[:]
    #assert not ( len( paramsRest ) % 3 ) #0,1,2면  assert error 발생
    #return off + sum( [ lorentzian( x, *paramsRest[ i : i+3 ] ) for i in range( 0, len( paramsRest ), 3 ) ] )
    return sum( [ lorentzian( wavenumbers, *paramsRest[j][ : ] ) for j in range( 0, len( paramsRest ) ) ] )  
```
